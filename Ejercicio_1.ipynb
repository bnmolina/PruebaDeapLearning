```python
import numpy as np
def step(z):
    return 1 if z >= 0 else 0
def neurona(x1, x2, w1, w2, b):
    z = x1*w1 + x2*w2 + b
    y = step(z)
    return z, y
# Valores base
w1, w2, b = 1, 1, -2.0  # intenta parecerse a una compuerta AND
casos = [(0,0),(0,1),(1,0),(1,1)]
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b)
    print(f"Entrada=({x1},{x2})  z={z:.1f}  salida={y}")
```

R1:/ El valor con el que la neurona se hacerca mas al AND es el valor de -2.0

R2:/ El efecto que tiene subir o bajar las Bias en la salia es cambiar los valores de 1110 a 0001.
